```python
import pandas as pd
data0=pd.read_csv("file:///C:/Users/NAAZ%20FATIMA/Downloads/weatherHistory.csv")
data0.head()
#importing required dataset and module
```

```python
data0.isnull().sum()
#checking null-data
```

```python
data0=data0.drop(['Precip Type'],axis=1)
#removing  precip type column as there is no requiremnt of this column in our hypothesis and it contains null values
```

```python
data0 = data0.rename(columns={'Formatted Date': 'Date'})
data0.head()
```

```python
import datetime
data0.Date= pd.to_datetime(data0.Date ,utc=True)
data0.head()
#coverting date and time to datetimeindex using datetime module
```

```python
df = data0.resample('M', on='Date').mean()
df.shape
#resampling the  hourly data into monthly data using resample
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
#converting datetime index to column
```

```python
df.head()
```

```python
#df['Date'] = pd.to_datetime(df['Date'])
dr= df[df['Date'].dt.month == 1]
dr.shape
#acessing data for requried month
```

```python
import matplotlib.pyplot as plt
plt.ylabel('Apparent Temperature (C)')
plt.xlabel('Year')
plt.plot(dr['Date'],dr['Apparent Temperature (C)'],marker='.')
plt.show()
plt.ylabel('Humidity')
plt.xlabel('Year')
plt.plot(dr['Date'],dr['Humidity'],'g',marker='.')
plt.show()
#plotting the required curve and analysing effect of global warming on average apparent temperature and average humidity
```

From the above graphs we can infer that, For january average APPARENT TEMPERATURE and average HUMIDITY decreases-increases and 
vice-versa as per the given data and finally as compared to 2006 average APPARENT TEMPERATURE increases and average HUMIDITY increases 
in 2016 due to global warming.
